```julia
"""
Computes the effective area for nutau + nutaubar for a given (anti) neutrino energy
"""
function effective_area(energy:: Number, nmodules=22_000)
    les = log.([4.167419115058369e14, 7.410835605310925e14, 1.3178536368107755e15, 2.343511987785515e15, 4.167419115058369e15, 7.410835605310926e15, 1.317853636810779e16, 2.3435119877855244e16, 4.167419115058386e16, 7.410835605310955e16, 1.3178536368107811e17, 2.3435119877855245e17, 4.167419115058386e17, 7.410835605310956e17])
    laeffs = log.([3.7487906588945906e14, 7.157874704876956e15, 1.1171710673062234e16, 4.157390439419208e16, 1.0574667638600392e17, 2.8133374665457165e17, 4.8706048220581875e17, 1.4305098647937477e18, 3.07580824077941e18, 3.8746388874890563e18, 6.997841927298492e18, 1.0474857532157921e19, 1.1679066571702622e19, 1.3985795606407016e19])
    itp = linear_interpolation(les, laeffs)
    return exp(itp(log(energy))) * nmodules / 22_000
end
```

```julia
"""
Eq. 6 from https://arxiv.org/pdf/hep-ph/0208187.pdf
"""
function xbounds(Eν, mℓ, mn)::Tuple{Float64, Float64}
    return mℓ^2 / 2 / mn / (Eν - mℓ), 1.0
end

"""
Eqs. 7 and 8 from https://arxiv.org/pdf/hep-ph/0208187.pdf
"""
function ybounds(Eν, x, mℓ, mn)::Tuple{Float64, Float64}
    numa = 1 - mℓ^2 * ((1 / 2 / mn / Eν / x) + 1 / (2 * Eν^2))
    numb = √((1 - mℓ^2 / 2 / mn / Eν / x)^2 - mℓ^2 / Eν^2)
    denum = 2 * (1 + mn * x / 2 / Eν)
    a = numa / denum
    b = numb / denum
    
    return a - b, a + b
end
```

```julia
"""
Compute the cosine of the opening angle in DIS interaction
"""
function dis_cosθ(Eν::Number, x::Number, y::Number, mℓ::Number, mn::Number)
    
    coef = 1 / √(Eν^2 * (1 - y)^2 - mℓ^2) / Eν / 2
    t1 = 2 * Eν^2 * (1 - y) - 2 * x * y * Eν * mn - mℓ^2
    return coef * (t1)
end

function dis_cosθ(Eν::Number, x::Number, y::Number)
    return dis_cosθ(Eν::Number, x::Number, y::Number, mτ, mn)
end

"""
From Vol. 50:679-749 (Volume publication date December 2000) 
"""
function dis_cosθbar(Eν)
    return 0.7 * (Eν / units.TeV) ^ (-0.7)
    # return cos(deg2rad(0.7 * (Eν / units.TeV) ^ (-0.7)))
end
```

```julia
function wrapper(Eν, γ)
    return (Eν)^-γ * effective_area(Eν)
end

γ = 2.4
my_integral(le, p) = exp(le) * wrapper(exp(le), γ)
domain = (log(emin), log(emax))

prob = IntegralProblem(my_integral, domain)
sol = solve(prob, QuadGKJL())
```

```julia
Δt_tambo = 3 * units.second * 10^7.5
```

```julia
κ = 1 / sol.u / Δt_tambo
```

```julia
ic_effa_file = CSV.File("./resources/icecube_10year_ps/irfs/IC86_II_effectiveArea.csv")
ic_energies = []
ic_effas = []
for row in ic_effa_file
    if row[3] != 0.0
        continue
    end
    push!(ic_energies, (10^row[1] + 10^row[2]) * units.GeV / 2)
    push!(ic_effas, row[5] * units.cm^2)
end
```

```julia
function get_ic_effa(Eν)
    itp = linear_interpolation(log.(ic_energies), log.(ic_effas))
    return exp(itp(log(Eν)))
end
```

```julia
get_ic_effa(3units.TeV) / units.m^2
```

```julia
Δt_ic = 15units.second
```

```julia
function wrapper(Eν, γ)
    return (Eν)^-γ * effective_area(Eν)
end

γs = LinRange(1.9, 3.5, 20)
res = []
for γ in γs
    my_integral(le, p) = exp(le) * wrapper(exp(le), γ)
    domain = (log(emin), log(emax))

    prob = IntegralProblem(my_integral, domain)
    sol = solve(prob, QuadGKJL())

    κ = 1 / sol.u / Δt_tambo
    Φ(Eν) = κ * Eν ^ -γ

    q(le, p) = exp(le) * Φ(exp(le)) * get_ic_effa(exp(le)) * Δt_ic
    domain = (log(minimum(ic_energies)), log(maximum(ic_energies)))

    prob = IntegralProblem(q, domain)
    sol = solve(prob, QuadGKJL())
    push!(res, sol.u)
    println((γ, sol.u))
end
```
